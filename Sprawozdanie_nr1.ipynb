1. Celem zadania jest porównanie działania oraz wydajności różnych algorytmów sortowania poznanych podczas zajęć laboratoryjnych.
Należy przeanalizować zarówno teoretyczne właściwości algorytmów, jak i praktyczny czas ich działania dla różnych rozmiarów danych.

2. Napisz program w języku C++, który implementuje i porównuje następujące algorytmy sortowania:
- Sortowanie bąbelkowe (Bubble Sort)
- Dwukierunkowe sortowanie bąbelkowe (Cocktail Shaker Sort)
- Sortowanie przez wybór (Selection Sort)
- Sortowanie przez wstawianie (Insertion Sort)
- Sortowanie metodą Shella (Shell Sort)
- Sortowanie przez scalanie (Merge Sort)

Program powinien:
- Generować tablice liczb losowych o różnych rozmiarach.
- Sortować dane każdym z powyższych algorytmów.
- Mierzyć czas działania algorytmu przy użyciu biblioteki <chrono>.
- Wyświetlać lub zapisywać wyniki pomiarów.

3. Testy należy przeprowadzić dla co najmniej trzech rozmiarów tablic:

- 1000 elementów
- 5000 elementów
- 10000 elementów

Każdy algorytm powinien sortować tę samą tablicę danych (należy wcześniej skopiować tablicę).

4. Wymagania dotyczące programu

Program powinien:
- implementować wszystkie algorytmy w oddzielnych funkcjach,
- generować dane losowe,
- mierzyć czas sortowania,
- prezentować wyniki w czytelnej formie (np. tabela).

Przykładowy format wyniku:

| Algorytm       | 1000 | 5000 | 10000 |
| -------------- | ---- | ---- | ----- |
| Bubble Sort    |      |      |       |
| Cocktail Sort  |      |      |       |
| Selection Sort |      |      |       |
| Insertion Sort |      |      |       |
| Shell Sort     |      |      |       |
| Merge Sort     |      |      |       |


5. Sprawozdanie

Na podstawie wyników programu przygotuj sprawozdanie, które powinno zawierać:

 - Wstęp teoretyczny czyli krótki opis każdego algorytmu: zasada działania, złożoność czasowa, złożoność pamięciowa.

 - Opis implementacji (sposobu generowania danych, metody pomiaru czasu, struktury programu.)

 - Wyniki eksperymentów czyli tabela z wynikami pomiarów czasu dla poszczególnych algorytmów.

 - Analiza wyników. Należy odpowiedzieć na pytania: Który algorytm był najszybszy? Który był najwolniejszy? Jak zmienia się czas działania wraz ze wzrostem liczby elementów? Czy wyniki są zgodne z teoretyczną złożonością algorytmów?

 - Wnioski. Krótka interpretacja wyników oraz podsumowanie porównania algorytmów.
